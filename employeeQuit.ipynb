```python
# now i am gonna apply smote to the training data only
# and based on my research this method gonna generate  samples for the minority class which are employees who quit
# so this helps balance the dataset and it's gonna make sure the model does not favor the majority class
smote = SMOTE(sampling_strategy=0.9, random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```
